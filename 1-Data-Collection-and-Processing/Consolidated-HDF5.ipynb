# Consolidated HDF5 Object

- TCGA Metadata
- GTEx Metadata

```python
import pandas as pd

# Synapse ID: syn12117384
exp = pd.read_hdf('/mnt/data/Expression/tcga_gtex_tpm.hd5')

# Synapse ID: syn12117386
tcga_path = 'Metadata-collation/output/tcga-metadata-cleaned.tsv'
tcga_met = pd.read_csv(tcga_path, sep='\t', index_col=0)

# Synapse ID: syn12117387
gtex_path = 'Metadata-collation/input_data/gtex-sample-metadata.tsv'
gtex_met = pd.read_csv(gtex_path, sep='\t', index_col=0)

# Synapse ID: syn11363589
met_path = '/mnt/data/Metadata/tcga_gtex_metadata_intersect.tsv'
met = pd.read_csv(met_path, index_col=0, sep='\t')

# Synapse ID: syn12117759
drug_path = '/mnt/data/Metadata/drug_biomarkers.tsv'
drug = pd.read_csv(drug_path, index_col=0, sep='\t')

# Synapse ID: syn12246910
pairing = pd.read_csv('data/subtype-pairings.tsv', sep='\t', index_col=0)
```

# Write Consolidated HDF

**Key: Value**
- exp: TCGA and GTEX gene expression (TPM)
- tcga: TCGA Metadata
- gtex: GTEx Metadata
- met: TCGA/GTEx intersectional metadata
- drug: Antineoplastic biomarkers and more

```python
output_hdf = '/mnt/data/Objects/tcga_gtex_data.hd5'
```

```python
exp.to_hdf(output_hdf, key='exp')
tcga_met.to_hdf(output_hdf, key='tcga')
gtex_met.to_hdf(output_hdf, key='gtex')
met.to_hdf(output_hdf, key='met')
drug.to_hdf(output_hdf, key='drug')
pairing.to_hdf(output_hdf, key='pairing')
```
